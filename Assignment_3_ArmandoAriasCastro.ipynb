```python
#Import libraries
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications import imagenet_utils
```

```python
#Load an image to classify
file_name = 'dog.jpg'
imag = image.load_img(file_name, target_size = (224, 224))
imag
```

```python
#Pre-process the image
resized_img = image.img_to_array(imag)
expanded_img = np.expand_dims(resized_img, axis = 0)
final_image = tf.keras.applications.mobilenet.preprocess_input(expanded_img)
```

```python
#Create model 1 -> MobileNet
mobile_net = tf.keras.applications.mobilenet.MobileNet()

#Predict with model 1
#Actual classification -> German Shepherd dog
prediction_1 = mobile_net.predict(final_image)
output_1 = imagenet_utils.decode_predictions(prediction_1)
print(output_1)
```

```python
#Create model 2 -> MobileNetV2
mobile_net2 = tf.keras.applications.mobilenet_v2.MobileNetV2()

#Predict with model 2
#Actual classification -> German Shepherd dog
prediction_2 = mobile_net2.predict(final_image)
output_2 = imagenet_utils.decode_predictions(prediction_2)
print(output_2)
```

```python
#Print top predictions for each model
print('Top predictions')
print('MobileNet predicts: ',output_1[0][0][1], ' with accuracy: ',  output_1[0][0][2])
print('MobileNetV2 predicts: ',output_2[0][0][1], ' with accuracy: ',  output_2[0][0][2])
```
